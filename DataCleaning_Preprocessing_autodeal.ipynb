```python
import pandas as pd
import numpy as np

df1 = pd.read_csv('/content/usedcars_autodeal_1.csv')
df2 = pd.read_csv('/content/usedcars_autodeal_2.csv')
```

```python
# The two dfs should have the same shape
print(df1.shape)
print(df2.shape)
```

```python
# Filling the N/A from df1 with df2, and assigning it to df
df = df1.fillna(df2)

df.head(-10)
```

```python
# Let's see how many N/As there are in the new df
df.isna().sum()
```

```python
df.shape
```

```python
df = df.drop_duplicates()
```

```python
df.shape
```

```python
# Removing the Year from Car Title and creating a new column for Year
df[['Year', 'Car Title']] = df['Car Title'].str.split(n=1, expand=True)

df.head(10)
```

```python
# Creating column containing the cars brand
df['Car Brand'] = df['Car Title'].apply(lambda x: x.split(' ', 1)[0].strip())

df.head(10)
```

```python
df['Car Mileage'] = df['Car Mileage'].str.rstrip(' Km')
df['Car Price'] = df['Car Price'].str.lstrip('P')

df.head()
```

```python
df['Car Mileage'] = df['Car Mileage'].str.replace(',', '', regex=True).astype('Int64')
df['Car Price'] = df['Car Price'].str.replace(',', '', regex=True).astype('Int64')
```

```python
df.head(10)
```

```python
pd.set_option('display.max_rows', 10)

name_column = df['Car Title']
print(name_column)
print("\n")
```

```python
# Extract alphanumeric and space characters from the 'Car Title' column to create the 'Car Name' column.
df['Car Name'] = df['Car Title'].str.extract('([a-zA-Z0-9 -]+)')

# Display the first 10 rows of the DataFrame.
df.head(10)
```

```python
# # Remove 'AT', 'MT', or 'CVT' from car Name
# df['Car Name'] = df['Car Name'].str.replace(r'\s(AT|A/T|MT|M/T|CVT)', '', regex=True)
```

```python
# Extract the first two words from the 'Car Name' column and join them to create a new column.
df['Car Name'] = df['Car Name'].str.split(n=2).str[:2].str.join(' ')
```

```python
name_column = df['Car Name']
print(name_column)
print("\n")
```
